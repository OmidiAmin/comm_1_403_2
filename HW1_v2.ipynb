```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(-10,10,0.001)
```

```python
N = 100
total = np.zeros_like(t)
for n in range(-N,N,1):
  if n %2 == 0:
    pass
  else:
    total = total + 1/(2**0.5 * n**2)  * np.exp(1j * 2 * np.pi * n * t)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def Delta_fun(x):
    return (1 - np.abs(x))*(np.abs(x) <= 1)
```

```python
t= np.arange(-3,3,0.1)
plt.plot(t, Delta_fun(t))
```

```python
t= np.arange(-10,10,0.01)
x = np.zeros_like(t)
N = 5
for n in range(-N-1,N+1,1):
  x += Delta_fun(t - 2*n)
plt.plot(t, x)
```

```python
def c_k(k):
  if k ==0:
    return 0.5
  else:
    return 2/(k *np.pi)**2 * np.sin(k * np.pi /2) ** 2
```

```python
c_k(5)
```

```python
n_vec = np.arange(-10,10,1)
x_n_vec = np.zeros(len(n_vec))
for n_idx, n in enumerate(n_vec):
  x_n_vec[n_idx] = c_k(n)
plt.stem(n_vec, np.abs(x_n_vec), 'o')
```

```python
t= np.arange(-1,3,0.001)
x = np.zeros_like(t)
N = 5
for n in range(-N-1,N+1,1):
  x += Delta_fun(t - 2*n)
```

```python
N = 0
T = 2
n_vec = [0]
x_reconstruc = np.zeros(len(t)).astype(np.complex128)
for n_idx, n in enumerate(n_vec):
  x_reconstruc += c_k(n) * np.exp(1j * 2 * np.pi * n * t /T)
plt.plot(t, x, label= 'Original')
plt.plot(t, x_reconstruc.real, label= 'reconstructed')
plt.legend()
plt.title(f"N: {N}")
```
