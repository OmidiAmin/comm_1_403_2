```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def R_X(m,x,M):
  N = len(x) - 1
  RX = 0
  if (m <= M) and (m >= 0):
    for n in range(1, N - m + 1,1):
      RX += x[n]*x[n+m]
    RX = (RX / (N -m))
  elif (m <= -1) and (m >= -M):
    for n in range(-m, N + 1,1):
      RX += x[n]*x[n+m]
    RX = (RX / (N +m))
  return RX

def S_X(f, x, M):
  sum_ = 0
  for m in range(-M,M+1,1):
    sum_ += R_X(m,x,M) * np.exp(- 1j * 2 * np.pi * f * m / (2*M+1))
    return sum_
def x_gen(N):
  w= np.random.randn()
  x = np.zeros(N+1)
  x[0] = w
  for n in range(1, N+1,1):
    x[n] = x[n-1] + w
  return x
```

```python
N = 1000
M = 100
x= x_gen(N)
Rx_vec = np.zeros(2*M + 1)
m_vec = np.arange(-M,M+1,1)
f_vec = np.arange(-5,5,0.1)
Sx_vec = np.zeros(len(f_vec)).astype(np.cdouble)
for m_idx in range(len(m_vec)):
  m = m_vec[m_idx]
  Rx_vec[m_idx] = R_X(m,x,M)

for f_idx in range(len(f_vec)):
  f = f_vec[f_idx]
  Sx_vec[f_idx] = S_X(f, x, M)
```

```python
plt.figure()
plt.plot(m_vec, Rx_vec, '*')
plt.xlabel('$m$')
plt.ylabel('$R_x(m)$')
plt.figure()
plt.plot(f_vec, Sx_vec.real, '-*')
plt.xlabel('$f$')
plt.ylabel('real part of $S_x(f)$')
plt.figure()
plt.plot(f_vec, Sx_vec.imag, '-*')
plt.xlabel('$f$')
plt.ylabel('imaginary part of $S_x(f)$')
plt.figure()
plt.plot(f_vec, np.abs(Sx_vec), '-*')
plt.xlabel('$f$')
plt.ylabel('$|S_x(f)|$')
```
